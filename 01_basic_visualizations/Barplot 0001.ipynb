```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from datetime import datetime

current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
print("Lbraries Loaded Successfully")
```

```python
import pandas as pd

# Tourist visitors to different Tanzanian national parks (in thousands)
national_parks_visitors = {
    'Serengeti': 150, 'Ngorongoro': 120, 
    'Tarangire': 85, 'Lake Manyara': 65, 'Mikumi': 55,
    'Ruaha': 40, 'Selous': 35, 'Katavi': 25, 
    'Gombe': 15, 'Mahale': 10, 'Saadani': 8,
    'Udzungwa': 12, 'Kitulo': 5, 'Mkomazi': 7, 
    'Rubondo': 6, 'Burigi-Chato': 4, 'Ibanda-Kyerwa': 3,
    'Rumanyika-Karagwe': 2
}

df = pd.DataFrame(national_parks_visitors.items(), columns=['National Park', 'Visitors (Thousands)'])
df
```

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.rcParams.update({'font.family': 'Candara', 'font.style': 'normal', 'font.size': 11 })
bars = plt.bar(df['National Park'], df['Visitors (Thousands)'])

ax.set_title('Tourist Visitors to Tanzanian National Parks (in Thousands) - 2025', fontsize=12, fontweight='bold', pad=20)

# Add footer text
footer_text = "Plot by: Jabulente | Data Source: Dummy Dataset | Data-Driven Insights for a Better Future"
plt.text(0, -0.07, footer_text, ha='left', va='center', fontsize=10, color='black', transform=plt.gca().transAxes)
plt.text(0.8, -0.07, f"{current_time}", ha='left', va='center', fontsize=9, color='gray', transform=plt.gca().transAxes)


plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.savefig("../Visualizations/Tanzania National Parks 00001", dpi=600)
plt.show()
```

```python
def bar_highlight(value, average_value, good_value):
    if value >= good_value:
        return '#1b9e77'
    elif value >= average_value:
        return '#d95f02'
    else:
        return '#7570b3'

#cutoff values
good = 65
average = 10

df['colours'] = df['Visitors (Thousands)'].apply(bar_highlight, args=(average, good))

df = df.sort_values(by='Visitors (Thousands)')

fig, ax = plt.subplots(figsize=(8,8))
plt.rcParams.update({'font.family': 'Dubai', 'font.style': 'normal', 'font.size': 11 })
bars = plt.barh(df['National Park'], df['Visitors (Thousands)'], height=0.8, color=df['colours'])

ax.set_title('Tourist Visitors to Tanzanian National Parks (in Thousands) - 2025', fontsize=12, fontweight='bold', pad=20)

ax.spines[['right', 'top', 'bottom']].set_visible(False) 
ax.xaxis.set_visible(False)

ax.bar_label(bars, padding=1, color='black', fontsize=10, label_type='edge', fmt='%.0fK', fontweight='bold')

ax.axvline(x=65, zorder=0, color='grey', ls='--', lw=1.5)
ax.text(x=65, y=18, s='Most Visited National Parks', ha='center', fontsize=12, bbox=dict(facecolor='white', edgecolor='grey', ls=' '))

# Add footer text
footer_text = "Plot by: Jabulente | Data Source: Dummy Dataset | Data-Driven Insights for a Better Future"
plt.text(-0.17, -0.07, footer_text, ha='left', va='center', fontsize=10, color='black', transform=plt.gca().transAxes)
plt.text(0.8, -0.07, f"{current_time}", ha='left', va='center', fontsize=9, color='gray', transform=plt.gca().transAxes)


plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.savefig("../Visualizations/Tanzania National Parks 00009", dpi=600)
plt.show()
```
